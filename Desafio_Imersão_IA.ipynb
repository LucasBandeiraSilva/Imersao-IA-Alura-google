## Importações necessárias para o desevolvimento

```python
import numpy as np;
import pandas as pd
import google.generativeai as genai
GOOGLE_API_KEY="COLE A API KEY AQUI"
genai.configure(api_key=GOOGLE_API_KEY)
```

## Lendo arquivo Excel

```python
data = pd.read_excel("/content/Base_De_Dados_Games.xlsx")
```

## Criando o DataFrame

```python
df = pd.DataFrame(data)
```

```python
df.rename(columns={"Data de lançamento": "DataLancamento", "Avaliação Media" : "AvaliacaoMedia", "Mecanicas do jogo" : "Mecanicas"},inplace=True)
```

## Visualizando modelos para embedding

## Criando função de embedding

```python
model = "models/embedding-001"
def embedding_fn(titulo,genero,plataforma,dataLancamento,desenvolvedora,avaliacaoMedia,sinopse,mecanicas,personagens):
  texto_combinado = f"{titulo} {genero} {plataforma} {dataLancamento} {desenvolvedora} {avaliacaoMedia} {sinopse} {mecanicas} {personagens}"
  resultadoEmbedding = genai.embed_content(model=model, content=texto_combinado)["embedding"]
  return resultadoEmbedding

```

```python
df["embeddings"] = df.apply(lambda row: embedding_fn(row["titulo"],row["genero"],
                                                     row["plataforma"],row["DataLancamento"],
                                                     row["Desenvolvedora"],row["avaliacaoMedia"],
                                                     row["Sinopse"],row["mecanicas"],
                                                     row["Personagens"]),axis=1)
df
```

## Gerando consulta

```python
def consultaDados(consulta,dataFrame,model):
  consultaEmbedding = genai.embed_content(model=model,content=consulta,task_type="RETRIEVAL_QUERY")
  produtos_escalares = np.dot(np.stack(df["embeddings"]),consultaEmbedding["embedding"])
  indice = np.argmax(produtos_escalares)
  return df.iloc[indice]["titulo"]
  df
```

## Realizando a consulta

```python
consulta = "ragnar"
resultado = consultaDados(consulta=consulta,dataFrame=df,model=model)
print(resultado)
```
